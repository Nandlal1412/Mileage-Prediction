# **Import Library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Import Data**

```python
df=pd.read_csv('/content/auto-mpg.csv')
```

```python
df.head()
```

```python
df.nunique()
```

# **Data Processing**

```python
df.info()
```

```python
df.describe()
```

```python
df['horsepower'].unique()
```

```python
df['horsepower']=df['horsepower'].str.replace('?','70')
```

```python
df['horsepower']=df['horsepower'].astype(int)
```

```python
df.corr()
```

# **Removing Missing Values**

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.info()
```

# **Data Visualization**

```python
sns.pairplot(df,x_vars=['displacement','horsepower','weight','acceleration','mpg'],y_vars=['mpg'])
```

```python
sns.regplot(x='displacement',y='mpg',data=df)
```

# **Define Target Variable y and Feature X**

```python
df.columns
```

```python
y=df['mpg']
```

```python
y.shape
```

```python
X=df[['displacement','horsepower','weight','acceleration']]
```

```python
X.shape
```

```python
X
```

# **Scaling Data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X=scaler.fit_transform(X)
```

```python
X
```

```python
pd.DataFrame(X).describe()
```

# *After Standardization mean=1 and standard deviaion =0*

# **Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# **Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

**Mileage = 23.51 - 1.07 displacement - 1.40 horsepower - 4.45 weight - 0.25 acceleration**

# **Predict Test Data**

```python
y_pred=regressor.predict(X_test)
```

```python
y_pred
```

# **Model Accuracy**

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mse)
r_squared=r2_score(y_test,y_pred)
```

```python
print(f'Mean Squared Error is: {mse}')
print(f'Mean Absolute Error is: {mae}')
print(f'Root Mean Square Error is: {rmse}')
print(f'RSquared is: {r_squared}')
```
